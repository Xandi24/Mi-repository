```python
import numpy as np
import matplotlib.pyplot as plt
import time

def make_list(a):
    x=[0.1]
    for i in range (0,4999):
        x.append(a*x[i]*(1-x[i]))
    return x

def tcalc_med(x):
    ti=time.time()
    tam=len(x)
    s=0
    i=0
    while(i<tam):
        s=s+x[i]
        i=i+1
    R=s/tam
    tf= time.time() - ti
    return tf
    

def tcalc_var(x):
    ti=time.time()
    tam=len(x)
    s=0
    i=0
    m=calc_med(x)
    while(i<tam):
        s=s+(x[i]-m)**2
        i=i+1
    R=s/(tam-1)
    tf= time.time() - ti
    return (tf)

def calc_med(x):
    tam=len(x)
    s=0
    i=0
    while(i<tam):
        s=s+x[i]
        i=i+1
    return (s/tam)

def calc_var(x):
    ti=time.time()
    tam=len(x)
    s=0
    i=0
    m=calc_med(x)
    while(i<tam):
        s=s+(x[i]-m)**2
        i=i+1
    R=s/(tam-1)
    return R
```

## Gráficos de média

```python
x=np.linspace(1,40,40)*(0.1)
y=np.array([])

for i in range(1,41):
    y=np.append(y,tcalc_med(make_list(i*(0.1))))

plt.plot(x,y)
plt.grid()
plt.xlabel('Valor do termo "a"')
plt.ylabel('t[s]')
plt.title('Tempo das medias em funcao de "a"')
plt.show()
```

```python
x=np.linspace(1,40,40)*(0.1)
y=np.array([])

for i in range(1,41):
    y=np.append(y,calc_med(make_list(i*(0.1))))

plt.plot(x,y)
plt.grid()
plt.xlabel('Valor do termo "a"')
plt.ylabel('Valores resultantes')
plt.title('Valores das medias em funcao de "a"')
plt.show()
```

## Gráficos de  Variância

```python
x=np.linspace(1,40,40)*(0.1)
y=np.array([])

for i in range(1,41):
    y=np.append(y,tcalc_var(make_list(i*(0.1))))

plt.plot(x,y)
plt.grid()
plt.xlabel('Valor do termo "a"')
plt.ylabel('t[s]')
plt.title('Tempo das variancias em funcao de "a"')
plt.show()
```
