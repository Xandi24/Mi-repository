```python
import numpy as np
import matplotlib.pyplot as plt

n=10*7
rng=400000
red=0.00001
def make_list(a,N):
    x=[0.1]
    for i in range (0,N):
        x.append(a*x[i]*(1-x[i]))
    return x

x,y=[],[]

for i in range (0,rng+1):
    x.append(i*red)
    y.append(make_list(i*red,n))

plt.plot(x,y, ls='',marker=',')
plt.grid()
plt.xlabel('Valores de "a"')
plt.ylabel('Possiveis valores da sequencia')
plt.title('Grafico de bifurcacoes')
plt.show()
```
